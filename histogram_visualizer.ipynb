```python
import cv2 
from matplotlib import pyplot as plt 
```

```python
Path = '/home/locus/Cloud/NextCloud_SDU/Master/workdir/images/image_splitter sample/'

img_ch1 = cv2.imread(Path + 'image_241_ch1.jpg',0) 
img_ch2 = cv2.imread(Path + 'image_241_ch2.jpg',0) 
img_ch3 = cv2.imread(Path + 'image_241_ch3.jpg',0) 
img_ch4 = cv2.imread(Path + 'image_241_ch4.jpg',0) 
img_ch5 = cv2.imread(Path + 'image_241_ch5.jpg',0) 
img_ch6 = cv2.imread(Path + 'image_241_ch6.jpg',0) 

img_ch1_eq = cv2.imread(Path + 'image_241_ch1_eq.jpg',0) 
img_ch2_eq = cv2.imread(Path + 'image_241_ch2_eq.jpg',0) 
img_ch3_eq = cv2.imread(Path + 'image_241_ch3_eq.jpg',0) 
img_ch4_eq = cv2.imread(Path + 'image_241_ch4_eq.jpg',0) 
img_ch5_eq = cv2.imread(Path + 'image_241_ch5_eq.jpg',0) 
img_ch6_eq = cv2.imread(Path + 'image_241_ch6_eq.jpg',0) 
```

```python
fig, axs = plt.subplots(2,6,figsize=(25,10))
fig.suptitle('Original and equalized histograms\nx = bins, y = pixels', fontsize=30)

axs[0, 0].hist(img_ch1.ravel(),256,[0,256]) 
axs[0, 1].hist(img_ch2.ravel(),256,[0,256]) 
axs[0, 2].hist(img_ch3.ravel(),256,[0,256]) 
axs[0, 3].hist(img_ch4.ravel(),256,[0,256]) 
axs[0, 4].hist(img_ch5.ravel(),256,[0,256]) 
axs[0, 5].hist(img_ch6.ravel(),256,[0,256]) 
axs[1, 0].hist(img_ch1_eq.ravel(),256,[0,256]) 
axs[1, 1].hist(img_ch2_eq.ravel(),256,[0,256]) 
axs[1, 2].hist(img_ch3_eq.ravel(),256,[0,256]) 
axs[1, 3].hist(img_ch4_eq.ravel(),256,[0,256]) 
axs[1, 4].hist(img_ch5_eq.ravel(),256,[0,256]) 
axs[1, 5].hist(img_ch6_eq.ravel(),256,[0,256]) 

plt.savefig(Path + 'histogram_plot.jpg')
plt.show() 
```

```python
fig, axs = plt.subplots(6,2,figsize=(10,25))
fig.suptitle('Original versus equalized histograms\nx = bins, y = pixels', fontsize=30)

axs[0, 0].hist(img_ch1.ravel(),256,[0,256]) 
axs[1, 0].hist(img_ch2.ravel(),256,[0,256]) 
axs[2, 0].hist(img_ch3.ravel(),256,[0,256]) 
axs[3, 0].hist(img_ch4.ravel(),256,[0,256]) 
axs[4, 0].hist(img_ch5.ravel(),256,[0,256]) 
axs[5, 0].hist(img_ch6.ravel(),256,[0,256]) 
axs[0, 1].hist(img_ch1_eq.ravel(),256,[0,256]) 
axs[1, 1].hist(img_ch2_eq.ravel(),256,[0,256]) 
axs[2, 1].hist(img_ch3_eq.ravel(),256,[0,256]) 
axs[3, 1].hist(img_ch4_eq.ravel(),256,[0,256]) 
axs[4, 1].hist(img_ch5_eq.ravel(),256,[0,256]) 
axs[5, 1].hist(img_ch6_eq.ravel(),256,[0,256]) 

plt.savefig(Path + 'histogram_plot_alt.jpg')
plt.show() 
```
